```python
import networkx as nx
import matplotlib.pyplot as plt
from schema import parse_json_file, School

# Load the JSON schema using the parse_json_file function
schema: School = parse_json_file('../scraper/University of Central Florida-2024-FALL.json.old')

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges to the graph
def add_nodes_and_edges_school(school: School):
    G.add_node(school.name, type='school')
    for professor in school.professors:
        prof_node = f"{professor.firstname} {professor.lastName}"
        G.add_node(prof_node, type='professor')
        G.add_edge(school.name, prof_node)
        for course_code, course in professor.courses.items():
            course_node = f"{course.name} ({course.code})"
            G.add_node(course_node, type='course')
            G.add_edge(prof_node, course_node)
        #for review in professor.reviews:
        #    review_node = f"Review {review.date}"
        #    G.add_node(review_node, type='review')
        #    G.add_edge(prof_node, review_node)

# Start adding nodes and edges from the root
add_nodes_and_edges_school(schema)

```

```python
# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, iterations=50)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", arrows=True)
plt.title("JSON Schema Graph Visualization")
plt.show()
```

```python
from pyvis.network import Network
nt = Network('500px', '500px')

nt.from_nx(G)
nt.toggle_physics(True)
nt.show('nx.html')
```
